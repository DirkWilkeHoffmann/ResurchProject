```python
#Attempt to merge datasets for convenience and reliabiltiy:

import pandas as pd

# Merge sales data with article data
merged_sales_data = pd.merge(sales_data, article_data, on='article_code', how='left')

# Merge inventory data with article data
merged_inventory_data = pd.merge(movements_data, article_data, on='article_code', how='left')

```

```python
#Data cleaning

# Handle missing values
merged_sales_data.fillna(0, inplace=True)
merged_inventory_data.fillna(0, inplace=True)

# Remove duplicates
merged_sales_data.drop_duplicates(inplace=True)
merged_inventory_data.drop_duplicates(inplace=True)
```

```python
import pandas as pd

# Assuming df1 is your sales data and df2 is your inventory data
# First, ensure the datetime columns are in datetime format
sales_data["ticket_end_time"] = pd.to_datetime(sales_data["ticket_end_time"])
movements_data["entry_timestamp"] = pd.to_datetime(movements_data["entry_timestamp"])

sales_data = sales_data.dropna(subset=["ticket_end_time"])
movements_data = movements_data.dropna(subset=["entry_timestamp"])

# Sort the dataframes by the datetime columns (important for merging method)
df1 = sales_data.sort_values("ticket_end_time")
df2 = merged_inventory_data.sort_values("entry_timestamp")

# Merge asof (nearest key merge)
merged_df = pd.merge_asof(df1, df2, left_on="ticket_end_time", right_on="entry_timestamp", direction="nearest", tolerance=pd.Timedelta("15 minutes"))

# Display the merged DataFrame
merged_df.head()
```

```python
print("Sales Data Summary:")
print(sales_data.describe())

print("\nArticle Data Summary:")
print(article_data.describe())

print("\nInventory Movements Data Summary:")
print(movements_data.describe())
```

```python
# Visualize sales patterns
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_data, x='date_key', y='sales_qty_base_uom', estimator='sum')
plt.title('Daily Sales Quantity')
plt.xlabel('Date')
plt.ylabel('Sales Quantity')
plt.show()
```

```python
# Visualize product distributions
plt.figure(figsize=(10, 6))
top_categories = sales_data.groupby('category_desc')['sales_qty_base_uom'].sum().nlargest(10)
sns.barplot(x=top_categories.values, y=top_categories.index, orient='h')
plt.title('Top Selling Product Categories')
plt.xlabel('Total Sales Quantity')
plt.ylabel('Category')
plt.show()
```

```python
# Visualize inventory movements
plt.figure(figsize=(10, 6))
sns.heatmap(movements_data.pivot_table(index='site_code', columns='movement_type', values='quantity_base_uom', aggfunc='sum'), cmap='coolwarm')
plt.title('Inventory Movement Types across Stores')
plt.xlabel('Movement Type')
plt.ylabel('Store')
plt.show()
```
